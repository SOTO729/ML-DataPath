```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
df=pd.read_csv('Salary_dataset.csv')
```

```python
data=df.iloc[:,1:]
```

```python
val=np.array(data.YearsExperience)
```

```python
import seaborn as sb
```

```python
arr=(val-val.mean())/val.std()
```

```python
sb.kdeplot(arr)
```

```python
np.where(abs(arr)>1.5)
```

```python
arr
```

```python
def outlier(arr,threshold):
    arr=np.array(arr)
    zscore=(arr-arr.mean())/arr.std()
    return np.where(abs(zscore)>threshold)
```

```python
val=data.columns
```

```python
out1=outlier(data[val[0]],1.5)
out2=outlier(data[val[1]],1.5)
```

```python
index=list(set((np.append(out1[0],out2[0]))))
```

```python
newdata=data.drop(index)
```

```python
sb.scatterplot(data=newdata,x='YearsExperience',y='Salary')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
col=newdata.columns
```

```python
x=np.array(newdata[col[0]]).reshape(-1,1)
y=np.array(newdata[col[1]]).reshape(-1,1)
```

```python
xtra,xtest,ytra,ytest=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
model.fit(xtra,ytra)
```

```python
model.score(xtest,ytest)
```
